```python
# Save pandas dataframe as csv file to S3 bucket

def save_df_to_s3(df, bucket, filename):
    csv_buffer = StringIO()
    df.to_csv(csv_buffer)
    s3_resource = boto3.resource(
        's3',
        aws_access_key_id=aws_access_key_id,
        aws_secret_access_key=aws_secret_access_key
    )
    s3_resource.Object(bucket, filename).put(Body=csv_buffer.getvalue())
```

```python
def get_busroutes(i):
    response = requests.get("http://datamall2.mytransport.sg/ltaodataservice/BusRoutes?$skip="+str(500*i), \
                            headers={'AccountKey': account_key, 'accept' : 'application/json'})

    json_busroutes = response.json()

    df_busroutes = pd.DataFrame()
    for i in json_busroutes['value']:
        row = pd.DataFrame.from_dict(i, orient='index').T
        df_busroutes = pd.concat([df_busroutes, row])
        
    return df_busroutes
```

```python
def get_busstops(i):
    response = requests.get("http://datamall2.mytransport.sg/ltaodataservice/BusStops?$skip="+str(500*i), \
                            headers={'AccountKey': account_key, 'accept' : 'application/json'})
    json_busstops = response.json()

    df_busstops = pd.DataFrame()
    for i in json_busstops['value']:
        row = pd.DataFrame.from_dict(i, orient='index').T
        df_busstops = pd.concat([df_busstops, row])
    
    return df_busstops
```

### Top 10 Bus Stops with the highest passenger volume (individual csv)
- Intuitively, these will be bus interchanges (which also count as bus stops)
- For each of these interchanges, only stops with trips that start at or end at the interchange are included.
- Data for use in visualization which color-codes ratio of departure/arrivals at each of these stop. A higher ratio indicates that there were more trips that departed from this stop for the interchange. 

```python
dfvol_wkday = df[df['DAY_TYPE'] == 'WEEKDAY']

top10_stops = busstop_grp.BusStopCode.tolist()[0:10]

for X in top10_stops:
    bus_vol = dfvol_wkday[(dfvol_wkday['ORIGIN_PT_CODE'].isin([X])) | (dfvol_wkday['DESTINATION_PT_CODE'].isin([X]))]\
        .groupby(['PATH'])['TOTAL_TRIPS'].sum().reset_index()
    bus_vol['ORIGIN_PT_CODE'] = bus_vol['PATH'].apply(lambda x: x.split('/')[0])
    bus_vol['DESTINATION_PT_CODE'] = bus_vol['PATH'].apply(lambda x: x.split('/')[1])
    
    bus_vol1 = bus_vol.drop(['DESTINATION_PT_CODE'], axis=1).rename(columns={"ORIGIN_PT_CODE": "BusStopCode"})
    bus_vol1['type'] = 'origin'
    bus_vol2 = bus_vol.drop(['ORIGIN_PT_CODE'], axis=1).rename(columns={"DESTINATION_PT_CODE": "BusStopCode"})
    bus_vol2['type'] = 'destination'
    bus_vol_new = bus_vol1.append(bus_vol2)
    bus_vol_new['BusStopCode'] = bus_vol_new['BusStopCode'].astype(int)
    bus_vol_new = pd.merge(bus_vol_new, df_busstops, on='BusStopCode', how='left')
    bus_vol_new = bus_vol_new.drop(['PATH', 'RoadName'], axis=1)
    busInt = pd.DataFrame(bus_vol_new[bus_vol_new['BusStopCode'] == int(X)].iloc[0, :]).T
    busInt['type'] = 'Bus Interchange'
    busInt['TOTAL_TRIPS'] = 20000
    bus_vol_new = bus_vol_new[bus_vol_new['BusStopCode'] != int(X)]
    bus_vol_new = pd.concat([bus_vol_new, busInt])
  
    bus_vol_new.to_csv("./dianaow.github.io/busroutes/data/bus_" + str(X) + ".csv", index=False)
```

### Traffic within districts

```python
df['ORIGIN_DISTRICT'] = df['ORIGIN_PT_CODE'].astype(str)
df['DESTINATION_DISTRICT'] = df['DESTINATION_PT_CODE'].astype(str)
df['ORIGIN_DISTRICT'] = [i[0] for i in df['ORIGIN_DISTRICT']]
df['DESTINATION_DISTRICT'] = [i[0] for i in df['DESTINATION_DISTRICT']]
```

```python
traffic = pd.DataFrame()
for X in range(1,10):
    total = df[(df['ORIGIN_DISTRICT'] == str(X)) | (df['DESTINATION_DISTRICT'] == str(X))]['TOTAL_TRIPS'].sum()
    within = df[(df['ORIGIN_DISTRICT'] == str(X)) & (df['DESTINATION_DISTRICT'] == str(X))]['TOTAL_TRIPS'].sum()
    incoming = df[(df['DESTINATION_DISTRICT'] == str(X))]['TOTAL_TRIPS'].sum()
    outgoing = df[(df['ORIGIN_DISTRICT'] == str(X))]['TOTAL_TRIPS'].sum()
    results = pd.DataFrame(data={"code": X, "peprc_in": (within/total)*100, "perc_out": (outgoing/total)*100, "perc_within": (incoming/total)*100 }, index=[0])
    traffic = pd.concat([traffic, results])
```

### Departures vs Arrivals
#### For all busstops, find the total number of trips departed-from (origin) and arrived-to (destination) at the particular stop for the HOUR 
- I think this is more insightful than finding ratio for the entire month

```python
df_ratio = pd.DataFrame()
for i in range(5,24,1):
    df_ratio = pd.DataFrame()
    tmp = df[(df['TIME_PER_HOUR'] == i) & (df['DAY_TYPE'] == 'WEEKDAY')]
    for X in df_busstops['BusStopCode'].tolist(): 
        incoming = tmp[(tmp['DESTINATION_PT_CODE'] == X)]['TOTAL_TRIPS'].sum()
        outgoing = tmp[(tmp['ORIGIN_PT_CODE'] == X)]['TOTAL_TRIPS'].sum()
        ratio = pd.DataFrame(data={'BusStopCode': X, 'RATIO': (outgoing/incoming) * 100, 'TOTAL': (outgoing+incoming) }, index=[0])
        df_ratio = pd.concat([df_ratio, ratio])
    df_ratio.to_csv("./dianaow.github.io/busroutes/data/origin_destination_" + str(i) + ".csv", index=False)
    
```

```python
import math

df_ratio1 = df_ratio[df_ratio['BusStopCode'] >= 10000]
df_ratio1 = pd.merge(df_ratio1, df_busstops, on='BusStopCode', how='left')
df_ratio1 = df_ratio1.drop(['RoadName'], axis=1)
df_ratio1 = df_ratio1.sort_values('RATIO', ascending=False) 
#df_ratio1['RATIO'].fillna("null", inplace=True)
#df_ratio1['RATIO'].replace(math.inf, 'infinity', inplace=True
#df_ratio1.to_csv("./dianaow.github.io/busroutes/data/departures_vs_arrivals.csv", index=False)
```
